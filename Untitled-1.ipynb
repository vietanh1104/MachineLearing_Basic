```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import csv
import pandas as pd
from pandas import read_csv
import seaborn as sns
```

```python
data = pd.read_csv('insurance.csv', sep = ',')
data.head().style.background_gradient(axis=0)
```

```python
data.isnull().sum()
```

```python
def Bmi_Classifier(row):
    if row['bmi'] < 18.5:
        return "underweight"
    elif row['bmi'] < 25:
        return "normal"
    elif row['bmi'] < 30:
        return "overweight"
    else:
        return "obese"

def Age_Classifier(row):
    if row['age'] < 17:
        return "child"
    elif row['age'] < 31:
        return "adult"
    elif row['age'] < 46:
        return "middle-aged"
    else:
        return "old-aged"

data['bmi_status'] = data.apply(Bmi_Classifier, axis=1)
data['age_status'] = data.apply(Age_Classifier, axis=1)

data.head().style.background_gradient(axis=0)
```

```python
print(data[data['smoker'] == 'no']['age'].count(),' non-smoker')
print(data[data['smoker'] == 'yes']['age'].count(),' smoker')
```

```python
smoker_data = data.groupby('smoker').mean().round(2)
smoker_data
```

```python
data2 = data[data['smoker'] == 'yes'][['sex', 'children', 'region', 'charges','bmi_status', 'age_status']]
new_data = pd.get_dummies(data2, columns = ['sex','children','region','bmi_status','age_status'])
new_data.head()
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
from sklearn.model_selection import train_test_split
```

```python
X = new_data.drop('charges', axis=1)
y = new_data['charges']
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=101)
```

```python
linear_model.fit(X_train, y_train)
```

```python
test_predictions = linear_model.predict(X_test)
```

```python
error = y_test - test_predictions
```

```python
plt.figure(dpi=100)
a = range(int(min(y)), int(max(y)), 1000)
sns.scatterplot( x = test_predictions, y = error )
plt.title("Test Result")
plt.xlabel("TEST_CHARGE PREDICTIONS")
plt.ylabel("Test Error")
plt.axhline(y=0,color = 'black', ls='--' )
```

```python
charge_predicton = linear_model.predict(X)
a = range(int(min(y)), int(max(y)), 1000)
b=a
plt.figure(figsize=(5,5),dpi=100)
plt.plot(a, b, color='r' )
sns.scatterplot(x=charge_predicton, y=y)
plt.xlabel("Predicted Charge")
plt.ylabel('Actual Charge')
```
